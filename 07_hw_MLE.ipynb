Для избавления от копипаста используйте функции. Копипаст является серезным источником ошибок на практике.

```python
# nan
df_chosen_columns = df_chosen_columns.dropna()

# выбросы
def find_outliers_iqr(data):
    Q1 = np.nanquantile(data, 0.25)
    Q3 = np.nanquantile(data, 0.75)
    IQR = Q3 - Q1
    outliers = data[(Q1 - 1.5*IQR > data)|(data > Q3 + 1.5*IQR)]
    return outliers
for col in columns:
    outliers = find_outliers_iqr(df_chosen_columns[col])
    df_chosen_columns = df_chosen_columns[~df_chosen_columns[col].isin(outliers)]
```

```python
# считаем ОМП
def mle(data, beta = None):
    if beta != None:
        theta = beta / np.mean(data)
        return theta
    else:
        mean = np.mean(data)
        variance = np.var(data)
        return mean, variance

pBRAF_N_mle = mle(df_chosen_columns['pBRAF_N'])
pCREB_N_mle = mle(df_chosen_columns['pCREB_N'])
ITSN1_N_mle = mle(df_chosen_columns['ITSN1_N'], 13.5)
DYRK1A_N_mle = mle(df_chosen_columns['DYRK1A_N'], 9)

print(f"pBRAF_N: ОМП (среднее, дисперсия) = {pBRAF_N_mle}")
print(f"pCREB_N: ОМП (среднее, дисперсия) = {pCREB_N_mle}")
print(f"ITSN1_N: ОМП (theta) = {ITSN1_N_mle}")
print(f"DYRK1A_N: ОМП (theta) = {DYRK1A_N_mle}")
```

Реализуйте функции, возвращающие значения функции правдоподобия  и логарифмической функции правдоподобия для гамма распределения с параметрам $\theta, \beta$. Функция принимает реализацию выборки и параметры. Логарифмическую функцию реализуйте с использованием формулы после ее упрощения, а не как взятие функции логарифма от уже вычисленного значения функции правдоподобия.

Методами `pdf` и `logpdf` пользоваться запрещено. Для подсчета гамма-функции используйте `scipy.special.gamma`, а для подсчета логарифма гамма-функции &mdash; `scipy.special.loggamma`. 


```python
def gamma_likelihood(data, theta, beta):
    likelihood = np.prod((theta ** beta) * (data ** (beta - 1)) * np.exp(-theta * data) / (gamma(beta)))
    return likelihood

def gamma_log_likelihood(data, theta, beta):
    log_likelihood = len(data)*(beta * np.log(theta) - loggamma(beta)) + np.sum((beta - 1) * np.log(data)) - theta * np.sum(data)
    return log_likelihood
```

Выведите следующие значения:
- функции правдоподобия;
- логарифма, взятого от функции правдоподобия;
- логарифмической функции правдоподобия

на следующих реализациях выборок:
- `np.ones(5) * 5` &mdash; 5 пятерок;
- `np.ones(500) * 5` &mdash; 500 пятерок.

Параметр $\beta$ положите равным 10, параметр $\theta$ положите равным 2.

Сделайте выводы. Какую функцию лучше вычислять на практике и почему?


```python
beta = 10
theta = 2

data1 = np.ones(5) * 5
data2 = np.ones(500) * 5

likelihood1 = gamma_likelihood(data1, theta, beta)
log_likelihood1 = gamma_log_likelihood(data1, theta, beta)
log_from_likelihood1 = np.log(likelihood1)

likelihood2 = gamma_likelihood(data2, theta, beta)
log_likelihood2 = gamma_log_likelihood(data2, theta, beta)
log_from_likelihood2 = np.log(likelihood2)

print("Выборка 1 (5 пятерок):")
print(f"Функция правдоподобия: {likelihood1}")
print(f"Логарифмическая функция правдоподобия: {log_likelihood1}")
print(f"Логарифм функции правдоподобия: {log_from_likelihood1}")

print("\nВыборка 2 (500 пятерок):")
print(f"Функция правдоподобия: {likelihood2}")
print(f"Логарифмическая функция правдоподобия: {log_likelihood2}")
print(f"Логарифм функции правдоподобия: {log_from_likelihood2}")
```

**Вывод**

Для больших выборок (в нашем случае 500 пятёрок) значение для функции правдоподобия - очень маленькое число, так как получается перемножением маленьких чисел. В целом, работать с очень маленькими числами неудобно, а в случае огромной выборки это может привести к переполнению. Так что на практике лучше использовать логарифмическую функцию, устойчивую к таким проблемам, потому что идёт сложение величин.

## Задача 5.

Смоделируем шум разными способами и посмотрим, как проявляется робастность оценок и как она связана с их асимптотической дисперсией. Будем сравнивать между собой:
- выборочное среднее, $\overline{X}$
- выборочную медиану, $\hat{\mu}$
- усеченное среднее, $\overline{X}_{\alpha}$ (фукнция `sps.trim_mean()`)

В качестве базовой модели будем рассматривать $N(\theta, 1)$, для которого нам нужно оценить параметр $\theta$. Однако, как бывает часто в жизни, нам приходят не чистые измерения, а с некоторым шумом, который может иметь разную природу. Чтобы его смоделировать, сгенерируем выборку из смешанного распределения следующим образом:
1. Сгенерируем выборку из нашего распределения $N(\theta, 1)$;
2. Сгенерируем выборки из распределения, которое задает шум (см. примеры ниже);
3. Сгенерируем монетку из распределения Бернулли, которая дает 1 с вероятностью $\varepsilon$, где $\varepsilon \in (0, 0.5)$; 
4. На тех позициях, где выпал 0, возьмем данные из $N(\theta, 1)$, где 1 - из шумовых данных.

Например,

```python
theta = 0
sample_count = 10000
sample_size = 100

# Создаем истинное распределение данных и шума
X_base = sps.norm(loc=theta).rvs(size=(sample_count, sample_size))
X_outlier = sps.norm(scale=3).rvs(size=(sample_count, sample_size))

# Создаем монетку, которая будет определять долю шума в данных
eps = 0.2
coin = sps.bernoulli(p=eps).rvs(size=sample_size)

X = np.where(coin, X_outlier, X_base)
X.shape
```
