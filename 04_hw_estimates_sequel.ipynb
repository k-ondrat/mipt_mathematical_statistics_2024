```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.5, palette='Set2')

from matplotlib.lines import Line2D
```

## Задача 5.

Пусть $X_1, \ldots, X_n$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$. Визуализируйте свойства состоятельности и асимптотической нормальности для оценки $S^2$ (см. задачу 1) дисперсии $\sigma^2$.


```python
sample_size = 300  # размер выборок
sample_count = 500  # количество выборок
n_range = (np.arange(sample_size) + 1)  # размеры подвыборок
```

Сгенерируем множество выборок из распределения $\mathcal{N}(a, \sigma^2)$ (количество выборок - `sample_count`, размер каждой выборки - `sample_size`)

```python
# генерируем множество выборок
samples = sps.norm.rvs(size=(sample_count,sample_size))
```

```python
# Вычислим оценки в зависимости от размера выборки
average_samples = np.mean(samples, axis = 0)
estimation = ((samples - average_samples) ** 2).cumsum(axis = 1)/ n_range
```

```python
# Добавим доверительный интервал

right = 1 + 2 / np.sqrt(np.arange(1, sample_size+1))
left = 1 - 2 / np.sqrt(np.arange(1, sample_size+1))
```

```python
def est_plot(estimation, sample_count, sample_size, left=None, right=None, xlim = sample_size ):
    '''
    estimation -- массив оценок от размера выборки
    sample_count -- количество генерируемых выборок
    sample_size -- размер каждой выборки
    left, rigth -- границы доверительного интеравала, нужны будут далее
    xlim = область по x
    '''
    
    plt.figure(figsize=(15, 7))
    for i in range(sample_count):
        plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
    if type(left) and type(right) is np.ndarray:
        
        plt.plot(np.arange(sample_size) + 1, left, color='black')
        plt.plot(np.arange(sample_size) + 1, right, color='black')
        labels = [r'$\hat{\theta}$', r'$\theta$',
                  'Границы доверительного интервала']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),
                   Line2D([0], [0], color='black', lw=2)]
        
    else: 
        labels = [r'$\hat{\theta}$', r'$\theta$']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),]
    
    plt.hlines(1, 0 ,sample_size, color='red')
    plt.title('Поведение оценки для разных реализаций')
    plt.xlabel('Размер выборки')
    plt.ylim((0, 5))
    plt.legend(handels, labels)
    plt.xlim((0, xlim));
```

```python
est_plot(estimation, sample_count, sample_size, left, right)
```

Визуализация подтверждает свойство *асимптотической нормальности*, так как большая часть значений оценки при больших размерах выборки отклоняется от истинного значения параметра не более чем на $2\theta\left/\sqrt{n}\right.$ (границы доверительного интервала).

Также, визуально оценка является состоятельной - значение сходится по вероятности к истинному значению.

## Задача 6.

Пусть $X_1,...,X_n$ выборка из из распределения $U(0, \theta)$. На этот раз будем работать с оценкой $\widehat{\theta} = (n + 1) X_{(1)}$. Визуализируйте свойства состоятельности и несмещенности данной оценки.

```python
theta=2
sample_size = 300  # размер выборок
sample_count = 500  # количество выборок
n_range = (np.arange(sample_size) + 1)  # размеры подвыборок
X = sps.uniform(loc = 0, scale = theta).rvs(size=(sample_count, sample_size))
```

***Указание:*** Для подсчета оценки воспользуйтесть  методом numpy `minimum.accumulate` (подробнее [тут](https://numpy.org/doc/stable/reference/generated/numpy.ufunc.accumulate.html) )

```python
estimation = (n_range + 1) * np.minimum.accumulate(X, axis=1)
```

```python
def est_plot(estimation, sample_count, sample_size, left=None, right=None, xlim = sample_size ):
    '''
    estimation -- массив оценок от размера выборки
    sample_count -- количество генерируемых выборок
    sample_size -- размер каждой выборки
    left, rigth -- границы доверительного интеравала, нужны будут далее
    xlim = область по x
    '''
    
    plt.figure(figsize=(15, 7))
    for i in range(sample_count):
        plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
    if type(left) and type(right) is np.ndarray:
        
        plt.plot(np.arange(sample_size) + 1, left, color='black')
        plt.plot(np.arange(sample_size) + 1, right, color='black')
        labels = [r'$\hat{\theta}$', r'$\theta$',
                  'Границы доверительного интервала']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),
                   Line2D([0], [0], color='black', lw=2)]
        
    else: 
        labels = [r'$\hat{\theta}$', r'$\theta$']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),]
    
    plt.hlines(theta, 0 ,sample_size, color='red')
    plt.title('Поведение оценки для разных реализаций')
    plt.xlabel('Размер выборки')
    plt.ylim((0, 5))
    plt.legend(handels, labels)
    plt.xlim((0, xlim));
```

```python
est_plot(estimation, sample_count, sample_size)
```
