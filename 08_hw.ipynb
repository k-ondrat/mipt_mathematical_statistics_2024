Построим для этих данных доверительные интервалы. Так как мы не знаем распределения, из которого нам приходят результаты эксперимента, воспользуемся бутстрепом.

**1.** Реализуйте подсчет массива бутстрепных статистик для данной выборки и вычисление трех доверительных интервалов.

```python
def bootstrap(sample, estimator, B_sample_size=None, B=10_000): 
    '''
    Строит бустрепные оценки по реализации выборки

    :param sample: выборка
    :param estimator: функция подсчета оценки
    :param B_sample_size: размер бустрепной выборки,
                          если None, то равен размеру исходной выборки
    :param B: число бустрепных выборок
    '''
    if B_sample_size is None:
        B_sample_size = len(sample)
        
    estimates = np.zeros(B)
    for i in range(B):
        bootstrap_sample = np.random.choice(sample, B_sample_size, replace=True)
        estimates[i] = estimator(bootstrap_sample)
    return estimates
```

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы 
    нормального бутстрепного доверителного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    std = np.std(bootstrap_estimations) 
    z = sps.norm.ppf((1 + alpha) / 2)
    return (theta - z * std).round(4), (theta + z * std).round(4)

def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы 
    центрального бутстрепного доверителного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    sorted_estimations = np.sort(bootstrap_estimations)
    l_index = round(len(sorted_estimations)*((1+alpha)/2))
    u_index = int(len(sorted_estimations)*((1-alpha)/2))

    l_bound = 2*theta - sorted_estimations[l_index]
    u_bound = 2*theta - sorted_estimations[u_index]

    return l_bound.round(4), u_bound.round(4)

def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Строит левую и правую границы 
    квантильного бутстрепного доверителного интервала

    :param theta: значение статистики на выборке
    :param bootstrap_estimations: бутстрепные статистики
    :param alpha: уровень доверия
    '''
    sorted_estimations = np.sort(bootstrap_estimations)
    l_index = int(len(sorted_estimations)*((1-alpha)/2))
    u_index = round(len(sorted_estimations)*((1+alpha)/2))
    l_bound = sorted_estimations[l_index]
    u_bound = sorted_estimations[u_index]

    return l_bound.round(4), u_bound.round(4)
```

**2.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = EX_1$. Результаты запишите в виде таблицы. Сделайте выводы.

```python
df_ci = pd.DataFrame(index = ['normal', 'central', 'quantile', 'theta'], 
                     columns=data.columns)

data = data.dropna()

for protein in df_ci.columns:
    theta = (np.array(data[protein])).mean()
    bootstrap_estimations = bootstrap(np.array(data[protein]), np.mean)
    df_ci[protein]['normal'] = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
    df_ci[protein]['central'] = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
    df_ci[protein]['quantile'] = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)
    df_ci[protein]['theta'] = theta

df_ci
```

## Задача 5

Бутстреп можно использовать для того, чтобы построить доверительные интервалы статистик, для которых у нас нет (или мы не знаем) хороших теорем об их распределениях. Рассмотрим коэффициент корреляции Пирсона.

$$r_{pearson}(X, Y) = \frac{\sum_{i=1}^n(X_i - \overline{X})(Y_i - \overline{Y})}{\sqrt{\sum_{i=1}^n(Y_i - \overline{Y})^2}\sqrt{\sum_{i=1}^n(Y_i - \overline{Y})^2}} = \frac{cov(X, Y)}{S_X \cdot S_Y},$$
где $X$ и $Y$ - векторы значений двух признаков длины $n$, $S$ - выборочные среднеквадратичные отклонения,  $cov$ - выборочная ковариация, $\overline{X}$ и $\overline{Y}$ - выборочные средние $X$ и $Y$. В питоне для подсчета выборочной корреляции можно использовать `sps.pearsonr()` или `np.corrcoef()`.

Чем ближе $r_{pearson}$ к $\pm1$, тем выше степень монотонной зависимости между признаками. На графике это выглядит так:

```python
sample_size = 500

X = sps.norm.rvs(size=sample_size)
Y = sps.norm.rvs(size=sample_size)
Z = 2 * X + sps.norm.rvs(size=sample_size)

r_weak = np.corrcoef(X, Y)[0, 1]
r_strong = np.corrcoef(X, Z)[0, 1]
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(21, 6))

ax = axs[0]
sns.regplot(x=X, y=Y,
            scatter_kws={"color": "black"}, 
            line_kws={"color": "red"}, 
            ci=None,
            ax=ax)

ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.set_title(f"Слабая корреляция r = {r_weak:.2f}")

ax = axs[1]
sns.regplot(x=X, y=Z,
            scatter_kws={"color": "black"}, 
            line_kws={"color": "red"}, 
            ci=None,
            ax=ax)

ax.set_ylabel("Z")
ax.set_xlabel("X")
ax.set_title(f"Сильная корреляция r = {r_strong:.2f}");
```

Реализуйте процедуру бутстрепа для корреляций между уровнями экспрессий каждой из пар белков: постройте гистограмму распределения бутстрепных коэффициентов корреляции и три вида доверительных интервалов. Результат оформите в виде графиков, сведенных в `plt.subplots()`, и общей таблицы со всеми доверительными интервалами (см. задание выше). Сделайте выводы.

```python
def pearson(sample1, sample2, B_sample_size=None, B=10_000): 
    if B_sample_size is None:
        B_sample_size = len(sample1)
        
    cr_pearson= np.zeros(B)
    for i in range(B):
        index_sample = np.random.choice(B_sample_size, B_sample_size)
        cr_pearson[i] = np.corrcoef(sample1[index_sample], sample2[index_sample])[0,1]
    return np.array(cr_pearson)
```

```python
fig, axes = plt.subplots(ncols = 3, figsize=(20, 7))
protein_par = [['ITSN1_N', 'ELK_N'], ['DYRK1A_N', 'ELK_N'], ['ITSN1_N', 'DYRK1A_N']]
results = pd.DataFrame(index = ['normal', 'central', 'quantile', 'theta'], columns = ['ITSN1_N vs ELK_N', 'DYRK1A_N vs ELK_N', 'ITSN1_N vs DYRK1A_N'])
for i in range(3):
        sample1 = data[protein_par[i][0]].values
        sample2 = data[protein_par[i][1]].values
        pearsons = pearson(sample1, sample2)
        theta = np.mean(pearsons)
        
        results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['normal']= bootstrap_normal_confidence_interval(theta, pearsons)
        results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['central'] = bootstrap_central_confidence_interval(theta, pearsons)
        results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['quantile'] = bootstrap_quantile_confidence_interval(theta, pearsons)
        results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['theta'] = theta

        sns.histplot(pearsons, ax=axes[i], bins=20)
        axes[i].axvline(theta, color='yellow', linestyle='-', label='Среднее')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['quantile'][0], color='red', label='Квантильный интервал')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['quantile'][1], color='red')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['normal'][0], color='green', label='Нормальный интервал')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['normal'][1], color='green')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['central'][0], color='blue', label='Центральный интервал')
        axes[i].axvline(results[f"{protein_par[i][0]} vs {protein_par[i][1]}"]['central'][1], color='blue')
        axes[i].set_title(f"{protein_par[i][0]} vs {protein_par[i][1]}")
        axes[i].legend()
plt.tight_layout()
plt.show()
```
